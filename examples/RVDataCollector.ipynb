```python
#from exotic.api.rv import RadialVelocityDatabase
import exotic.api.rv
from exotic.api.rv import RadialVelocityDatabase
from astroquery.simbad import Simbad
import csv
import os
import logging
logger = logging.getLogger("ultranest")
logger.addHandler(logging.NullHandler())
logger.setLevel(logging.WARNING)
import time
start_time = time.time()
# Target set-up
print()
print("Enter RadialVelocityDatabase")
print()              
print("Input the name of the stellar target. Do not include a planet designation.")
target = input("What is the name of your target? ")
simbad_result_count=0
while simbad_result_count == 0:
    try:
        result_table = Simbad.query_object(target)
        simbad_result_count = len(result_table)
    except:
        simbad_result_count = 0
        print("Target not found.")
        target = input("What is the name of your target? ") 
new_rv_data = RadialVelocityDatabase(target)
df = new_rv_data.get_data(target)

nt = len(df['BJD'])
print("nt in RVDATA ",nt)
print()
filetarget=target.replace(" ","")
print("filetarget in RVDATA ",filetarget)
print()
basepath = os.getcwd()
filepath = basepath + '/radial_velocity/data/'           
with open(filepath+filetarget+'.csv', 'w', newline='') as file:
    fieldnames = ['target', 'BJD','rv','rv_err','ins_name','RA','DEC','rjd']
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()
    i=0
    while i<nt:
        writer.writerow({"target": df["target"][i],"BJD":df["BJD"][i],"rv":df["rv"][i],
                        "rv_err":df["rv_err"][i],"ins_name":df["ins_name"][i],
                         "RA":df["RA"][i],"DEC":df["DEC"][i],"rjd":df["rjd"][i]})
        i+=1
end_time = time.time()
time_lapse = (end_time - start_time)/60
print(f'The time to completion was {time_lapse:.3f} mins.')
```
